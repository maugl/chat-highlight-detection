## Scipy peak predictor

```python
with open("../data/analysis/baselines/grid_search/eval_configs_20220218_10_25_26.json", "r") as in_file:
    baseline_scores = json.load(in_file)
```

```python
tot_scores = list()
cfgs = list()
for config in baseline_scores["spp"]:
    tot_scores.append(config["total_scores"])
    params = config["params"]["scipy_params"]
    params["shift"] = config["params"]["shift"]
    cfgs.append(params)
```

```python
df0 = pd.DataFrame.from_records(tot_scores)
df1 = pd.DataFrame.from_records(cfgs)
df = pd.concat([df0, df1], axis=1)
df = df[df["f-score"].apply(lambda x: len(x) < 3)]
```

```python
df["f_hl"] = df["f-score"].apply(lambda x: x[1])
```

```python
df.sort_values("f_hl", ascending=False)
```

## Message density as a classifier

```python
from baselines import load_experiments_data
from sklearn.metrics import precision_recall_curve, precision_recall_fscore_support
from sklearn.metrics import PrecisionRecallDisplay
import numpy as np
```

```python
matches = load_experiments_data("nalcs_*g[13]", load_random=None, random_state=None, data_path="./data")
```

```python
total_msg_density = list()
total_highlights = list()
for name, m in matches.items():
    total_msg_density.extend(m["cd_message_density_smoothed"])
    total_highlights.extend(m["highlights"])
```

```python
display = PrecisionRecallDisplay.from_predictions(total_highlights, total_msg_density, name="Smoothed message density")
_ = display.ax_.set_title("2-class Precision-Recall curve")
```

```python
precision, recall, thresholds = precision_recall_curve(total_highlights, total_msg_density)
```
