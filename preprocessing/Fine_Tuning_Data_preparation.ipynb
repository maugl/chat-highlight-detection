### Additional Signal for Highlight distance

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [15, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

```python
hl_test = chd.highlights[example_match]
```

```python
chd.get_match_meta_data(example_match)
```

```python
def get_decay(seq_len, max_decay_len, strength):
    # https://en.wikipedia.org/wiki/Exponential_decay
    decay_len = min(max_decay_len, seq_len)
    decay_seq = np.exp(-strength * np.linspace(0,1,num=decay_len))
    if max_decay_len < seq_len:
        return np.concatenate([decay_seq, np.zeros(seq_len - max_decay_len)])
    else:
        return decay_seq
```

```python
def generate_highlight_distances(hl_spans, vid_len, decay, max_decay_len, mode):
    assert mode in ["prev", "next"]
    last = 0
    hl_signal = list()
    for start, stop in hl_spans:
        decay_vals = get_decay(start - last, max_decay_len, decay)
        if mode == "prev":
            if last == 0:
                decay_vals = [0] * start
            hl_signal.extend(decay_vals)
        if mode == "next":
            hl_signal.extend(decay_vals[::-1])
        # add in length of highlight
        hl_signal.extend([1] * (stop - start))
        last = stop
    if mode == "prev":
        hl_signal.extend([0] * (vid_len - last))
    if mode == "next":
        hl_signal.extend(get_decay(vid_len - last, max_decay_len, decay))
    return hl_signal
```

```python
highlight_spans = chd.get_match_meta_data(example_match)["highlight_spans"]

hl_prev_exp = generate_highlight_distances(highlight_spans, len(hl_test), 6, 5000, "prev")
hl_next_exp = generate_highlight_distances(highlight_spans, len(hl_test), 10, 2000, "next")
```

```python
%matplotlib widget
plt.subplot(211)
plt.plot(list(range(len(hl_prev_exp))), hl_prev_exp, label="distance_previous", c="r")
plt.legend()
plt.subplot(212)
plt.plot(list(range(len(hl_prev_exp))), hl_next_exp, label="distance_next", c="b")
plt.legend()
plt.show()
```

```python
# discretize output to make it a classification problem
num_steps = 5

hldpd = np.floor(np.asarray(hl_prev_exp) * num_steps) / num_steps

hldnd = np.floor(np.asarray(hl_next_exp) * num_steps) / num_steps
```

```python
%matplotlib widget

plt.plot(range(len(hldpd)), hldpd, label="prev_discrete")
#plt.plot(range(len(hldnd)), hldnd, label="next_discrete")
plt.plot(range(len(hl_prev_exp)), hl_prev_exp, label="prev")
#plt.plot(range(len(hl_next_exp)), hl_next_exp, label="next")
plt.plot(range(len(hl_test)), hl_test, label="highlights")
plt.legend()
plt.show()
```

```python
def generate_highlight_durations(hl_spans, vid_len, decay, max_decay_len, mode):
    assert mode in ["start", "stop"]
    hl_signal = list()
    last = 0
    for start, stop in hl_spans:
        hl_signal.extend(np.zeros(start-last))
        decay_vals = get_decay(stop-start, max_decay_len, decay)
        if mode == "start":
            hl_signal.extend(decay_vals)
        if mode == "stop":
            hl_signal.extend(decay_vals[::-1])
        # add in length of highlight
        last = stop
    hl_signal.extend(np.zeros(vid_len-last))
    return hl_signal
```

```python
hl_start_exp = generate_highlight_durations(highlight_spans, len(hl_test), 7, 1000, "start")
hl_stop_exp = generate_highlight_durations(highlight_spans, len(hl_test), 7, 1000, "stop")
```

```python
%matplotlib widget

#plt.subplot(211)
plt.plot(list(range(len(hl_start_exp))), hl_start_exp, label="distance_start", c="r")
#plt.legend()
#plt.subplot(212)
plt.plot(list(range(len(hl_stop_exp))), hl_stop_exp, label="distance_stop", c="b")
plt.legend()
plt.show()
```
