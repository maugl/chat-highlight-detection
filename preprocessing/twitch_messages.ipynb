## Tokenization script

```python
def tokenize_messages(messages):
    tt = TweetTokenizer()
    return [tt.tokenize(msg) for msg in messages]

def tokenize_videos(path):
    chat_files = glob.glob(f"{path.strip('/')}/[0-9]*[0-9].txt")
    # print(len(chat_files))
    for cf in chat_files:
        with open(cf, "r") as in_file:
            lines = in_file.read().lower().split("\n")
            if lines:
                yield in_file.name.split("/")[-1].replace(".txt", ""), tokenize_messages(lines)
```

### Create message data info

```python
def load_tokenized_messages(path):
    chat_files = glob.glob(f"{path.strip('/')}/[0-9]*[0-9]_tok.txt")
    for cf in chat_files:
        with open(cf, "r") as in_file:
            lines = in_file.read().split("\n")
            lines_toks = [l.split(" ") for l in lines]
            if lines_toks and lines_toks[0]:
                yield in_file.name.split("/")[-1].replace("_tok.txt", ""), lines_toks
```

```python
msg_toks = iter(load_tokenized_messages("../data/videos_chat/corpus"))
```

```python
msg_data = dict()
token_freq = Counter()

i = 0
for vid, msgs in msg_toks:
    msg_data[vid] = {
            "num_tokens": 0,
            "num_chars": 0,
            "longest_message_toks": 0,
            "shortest_message_toks": math.inf,
            "longest_message_chars": 0,
            "shortest_message_chars": 0
    }
    mdat = msg_data[vid]
    for m in msgs:
        num_toks = len(m)
        num_chars = sum([len(c) for c in m]) # not including white space
        mdat["num_tokens"] = mdat["num_tokens"] + num_toks
        mdat["longest_message_toks"] = max(mdat["longest_message_toks"], num_toks)
        mdat["shortest_message_toks"] = min(mdat["shortest_message_toks"], num_toks)
        mdat["num_chars"] = mdat["num_chars"] + num_chars
        mdat["longest_message_chars"] = max(mdat["longest_message_chars"], num_chars)
        mdat["shortest_message_chars"] = min(mdat["shortest_message_chars"], num_chars)
        token_freq.update(Counter(m))
    """
    i+=1
    if i >= 10:
        break
    """
```

```python
# we know that @mentions are followed by a white space or end of line/message per twitch definition
re.sub("(^@[^\s]*\s|\s?@[^\s]*$|\s?@[^\s]*)", "", "@tag1 words1 words2 words3 @tag2 words4\n@tag3 words5")
```

```python
chat_files = glob.glob("../data/videos_chat/corpus/[0-9]*[0-9].txt")
# chat_files = glob.glob("../data/videos_chat/corpus/52706801.txt") # for testing


# https://urlregex.com/
mention_regex = re.compile(r"@[^\s]*")
bot_regex = re.compile(r"\n[!].*\n") # removing whole message with bot command
url_regex = re.compile(r'((http|ftp|https):\/\/)?([\w\-_]+(?:(?:\.[\w\-_]+)+))([\w\-\.,@?^=%&:/~\+#]*[\w\-\@?^=%&/~\+#])?') # matches anything that vaguely looks like a url
regex_space = re.compile(r"[^\S\r\n]{2,}")
regex_line = re.compile(r"\s{2,}")

for cf in chat_files:
    text_subbed = ""
    with open(cf, "r") as in_file:
        text = in_file.read()
        # remove @user_name mentions
        text_no_mentions = mention_regex.sub("", text)
        # remove lines with bot commands
        text_no_bot_commands = bot_regex.sub("\n", text_no_mentions)
        # remove urls
        text_no_urls = url_regex.sub("", text_no_bot_commands)
        # clean up multi white space
        text_no_dbl_spaces = regex_space.sub(" ", text_no_urls)
        # clean up multiple new lines
        text_no_dbl_line = regex_line.sub("\n", text_no_dbl_spaces)
        
        
    with open(cf, "w") as out_file:
        out_file.write(text_no_dbl_line)
```
